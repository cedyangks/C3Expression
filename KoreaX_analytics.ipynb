```python
DictWithKey:dict=dict.fromkeys([1,2,3,4,5]);
#setting up the dict with Function pointer
def AppendZero1(ArgStr:str)->str:
    print('AppendZero1');
    pass;
    LengthOfStr:int=len(ArgStr);
    ArgStr='0'+ArgStr;
    return ArgStr;

def AppendZero2(ArgStr:str)->str:
    print('AppendZero2');
    pass;
    LengthOfStr:int=len(ArgStr);
    ArgStr='00'+ArgStr;
    return ArgStr;
    
def AppendZero3(ArgStr:str)->str:
    print('AppendZero3');
    pass;
    LengthOfStr:int=len(ArgStr);
    ArgStr='000'+ArgStr;
    return ArgStr;

def AppendZero4(ArgStr:str)->str:
    print('AppendZero4');
    pass;
    LengthOfStr:int=len(ArgStr);
    ArgStr='0000'+ArgStr;
    return ArgStr;

def AppendZero5(ArgStr:str)->str:
    print('AppendZero5');
    pass;
    LengthOfStr:int=len(ArgStr);
    ArgStr='00000'+ArgStr;
    return ArgStr;
    
DictWithKey[1]=AppendZero1;
DictWithKey[2]=AppendZero2;
DictWithKey[3]=AppendZero3;
DictWithKey[4]=AppendZero4;
DictWithKey[5]=AppendZero5;

LoopCounter:int=0;
for i in KoreaSummary['Str Issue Code']:
    if (len(i)<6):
        KoreaSummary['Str Issue Code'][LoopCounter]=DictWithKey[6-len(i)](i);
        LoopCounter=LoopCounter+1;
        continue;
    LoopCounter=LoopCounter+1;
    continue;
```
