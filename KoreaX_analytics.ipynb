```python
def KRX_site()->webdriver.firefox.webdriver.WebDriver:
    #Query for certain stocks that are worth shorting
    BrowserKRX:list=[webdriver.firefox.webdriver.WebDriver] * 5;
    pass;
    BrowserKRX[0]=webdriver.Firefox();
    BrowserKRX[0].get('http://data.krx.co.kr/contents/MDC/MAIN/main/index.cmd');
    BrowserKRX[0].find_element(webdriver.common.by.By.XPATH,'/html/body/div[2]/section[1]/div[1]/div[2]/button').click();
    time.sleep(2);
    BrowserKRX[0].get('http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0203')
    time.sleep(2);
    print(type(BrowserKRX[0]));
    return BrowserKRX[0];
```
