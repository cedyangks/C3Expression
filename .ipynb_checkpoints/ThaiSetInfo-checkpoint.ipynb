```python
ThreadQueue=queue.Queue();

def WebQuery(WarrantTicker:str,ResultLists:list,Indexer:int) ->list:
    urlSite:str='https://www.set.or.th/en/market/product/stock/quote/'+WarrantTicker+'/price';
    FireBrowser:webdriver.firefox.webdriver.WebDriver=webdriver.Firefox();
    FireBrowser.get(urlSite);
    print(FireBrowser.title);
    WarrantInfo=FireBrowser.find_element(webdriver.common.by.By.XPATH,'//*[@id="stock-quote-tab-pane-1"]/div/div[1]/div/div[2]/div[2]/div');
    ResultLists[Indexer]:list=WarrantInfo.text.split('\n');
    FireBrowser.close();
    return ResultLists;

#IO operations better optimize with threading
Thread_ID_List:list=[None] * 11;
ParseContent_List:list=[None] * 11;
print(ParseContent_List)
urlList=['BRR-W1','BRR-W2','BANPU-W5','TTB-W1','NRF-W1','NEX-W2','APURE-W3','ASW-W2','BEYOND-W2','MINT-W9','ASW-W1','BTS-W7'];

#ThreadsList.append(WebQuery(urlList[1]));
#ThreadsList.append(WebQuery(urlList[2]));
#ThreadsList.append(WebQuery(urlList[3]));

#Spawn threads since it's a IO bound task
for i in range(11):
    Thread_ID_List[i]=threading.Thread(target=WebQuery,args=(urlList[i],ParseContent_List,i));
    Thread_ID_List[i].start();

for i in range(10):
    Thread_ID_List[i].join();

os.system("touch Thai_WarrantInfo.csv");
with open("Thai_WarrantInfo.csv",'w') as WarrantDetails:
        writer = csv.writer(WarrantDetails,delimiter=',',quoting=csv.QUOTE_MINIMAL);
        for EachList in ParseContent_List:
            writer.writerow(EachList);
    
```
